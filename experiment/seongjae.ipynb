```python
import pandas as pd

train =pd.read_csv('C:\\Users\\kimsj\\OneDrive\\바탕 화면\\대학교\\git\\AI-project\\data\\train.csv')
x_test =pd.read_csv('C:\\Users\\kimsj\\OneDrive\\바탕 화면\\대학교\\git\\AI-project\\data\\test.csv')

train
```

```python
from sklearn import preprocessing

def encode_features(dataDF):
    features = ["age_group","gender", "race", "religion"]
    for featue in features:
        le = preprocessing.LabelEncoder()
        le = le.fit(dataDF[featue])
        dataDF[featue] = le.transform(dataDF[featue])
        
    return dataDF

train = encode_features(train)
x_test = encode_features(x_test)

y_train = train["voted"]
x_train = train.drop('voted', axis=1)
```

```python
def get_categoty_time(Q_time):
    cat = 0
    if Q_time <= 500 : cat = 0
    elif Q_time <= 1000 : cat = 1
    elif Q_time <= 2000 : cat = 2
    elif Q_time <= 3000 : cat = 3
    elif Q_time <= 4000 : cat = 4
    elif Q_time <= 5000 : cat = 5
    elif Q_time <= 6000 : cat = 6
    elif Q_time <= 7000 : cat = 7
    elif Q_time <= 8000 : cat = 8
    elif Q_time <= 9000 : cat = 9
    elif Q_time <=10000 : cat = 10
    elif Q_time <=15000 : cat = 11
    elif Q_time <=20000 : cat = 12
    else : Q_time = 13

    return cat

Q_times = ['QaE', 'QbE', 'QcE', 'QdE', 'QeE', 'QfE', 'QgE', 'QhE', 'QiE', 'QjE', 'QkE','QlE', 'QmE', 'QnE', 'QoE', 'QpE', 'QqE', 'QrE', 'QsE', 'QtE']

for Q_time in Q_times:
    train[Q_time] = train[Q_time].apply(lambda x : get_categoty_time(x))
    x_test[Q_time] = x_test[Q_time].apply(lambda x : get_categoty_time(x))
```

```python
def get_categoty_age(age_num):
    num = age_num;
    if age_num == 0: num =7

    return num
train["age_group"] = train["age_group"].apply(lambda x : get_categoty_age(x))
x_test["age_group"] = x_test["age_group"].apply(lambda x : get_categoty_age(x))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score

knn_clf = KNeighborsClassifier(n_neighbors=4)
rf_clf = RandomForestClassifier(n_estimators=100, random_state=0)
dt_clf = DecisionTreeClassifier(random_state=0)
ada_clf = AdaBoostClassifier(n_estimators=100, random_state=0)
lr_clf = LogisticRegression(random_state=0)
gb_clf = GradientBoostingClassifier(random_state=0)
```

```python
import warnings
warnings.filterwarnings('ignore')

knn_clf.fit(x_train,y_train)
knn_pred = knn_clf.predict_proba(x_test)
```
