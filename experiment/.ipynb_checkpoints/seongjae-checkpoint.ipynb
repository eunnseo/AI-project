```python
import pandas as pd

train =pd.read_csv('C:\\Users\\kimsj\\OneDrive\\바탕 화면\\대학교\\git\\AI-project\\data\\train.csv')
test =pd.read_csv('C:\\Users\\kimsj\\OneDrive\\바탕 화면\\대학교\\git\\AI-project\\data\\test.csv')

train
```

```python
from sklearn import preprocessing

def encode_features(dataDF):
    features = ["age_group","gender", "race", "religion"]
    for featue in features:
        le = preprocessing.LabelEncoder()
        le = le.fit(dataDF[featue])
        dataDF[featue] = le.transform(dataDF[featue])
        
    return dataDF

train = encode_features(train)
test = encode_features(test)

y_train = train["voted"]
x_train = train.drop('voted', axis=1)
```

```python
def get_categoty_time(Q_time):
    cat = 0
    if Q_time <= 500 : cat = 0
    elif Q_time <= 1000 : cat = 1
    elif Q_time <= 2000 : cat = 2
    elif Q_time <= 3000 : cat = 3
    elif Q_time <= 4000 : cat = 4
    elif Q_time <= 5000 : cat = 5
    elif Q_time <= 6000 : cat = 6
    elif Q_time <= 7000 : cat = 7
    elif Q_time <= 8000 : cat = 8
    elif Q_time <= 9000 : cat = 9
    elif Q_time <=10000 : cat = 10
    elif Q_time <=15000 : cat = 11
    elif Q_time <=20000 : cat = 12
    else : Q_time = 13

    return cat


Q_times = ['QaE', 'QbE', 'QcE', 'QdE', 'QeE', 'QfE', 'QgE', 'QhE', 'QiE', 'QjE', 'QkE','QlE', 'QmE', 'QnE', 'QoE', 'QpE', 'QqE', 'QrE', 'QsE', 'QtE']
for Q_time in Q_times:
    train[Q_time] = train[Q_time].apply(lambda x : get_categoty_time(x))
    test[Q_time] = test[Q_time].apply(lambda x : get_categoty_time(x))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

dt_clf = DecisionTreeClassifier(random_state=0)
dt_clf.fit(x_train, y_train)
dt_pred = dt_clf.predict(x_train)
print("정확도{0:.4f}".format(accuracy_score(y_train,dt_pred)))
```

```python
dt_pred1 = dt_clf.predict(test)
print("정확도{0:.4f}".format(accuracy_score(y_train.sample(9107),dt_pred1)))
```

```python
from sklearn.ensemble import RandomForestClassifier
import warnings
warnings.filterwarnings('ignore')

rf_clf = RandomForestClassifier(random_state=0)
rf_clf.fit(test,y_train.sample(9107))
pred = rf_clf.predict(x_train)
print("랜덤 포렌스트 정확도: {0:.4f}".format(accuracy_score(y_train,pred)))
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score

xgb_clf = XGBClassifier(n_estimators=100, random_state=0)

xgb_clf.fit(x_train, y_train, early_stopping_rounds=10,
           eval_metric="auc", eval_set=[(x_train.sample(9107),y_train.sample(9107)),(test, y_train.sample(9107))])

xgb_roc_score = roc_auc_score(y_train, xgb_clf.predict_proba(x_train)[:,1],average="macro")
print("ROC_AUC: {0:.4f}".format(xgb_roc_score))
```
